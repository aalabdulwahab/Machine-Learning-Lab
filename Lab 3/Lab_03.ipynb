```python
df.drop(columns=['Car_ID']).describe()
df['Car_Age'] = 2026 - df['Manufacture_Year']
```

```python
df.drop(columns=['Car_ID', 'Manufacture_Year'], inplace=True)
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
##Distribution of the price
plt.figure(figsize=(10, 6))
sns.histplot(df['Price_USD'], bins=30, kde=True, color='blue')
plt.title("Distribution of Car Prices (USD)")
plt.xlabel("Price in USD")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(df['Car_Age'], bins=15, kde=True, color='green')


plt.title("Distribution of Car Age")
plt.xlabel("Age (Years)")
plt.ylabel("Frequency")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Car_Age', y='Price_USD',
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})

plt.title("Relationship between Car Age and Price")
plt.xlabel("Car Age (Years)")
plt.ylabel("Price (USD)")
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot with a regression line for Engine_CC
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Engine_CC', y='Price_USD',
            scatter_kws={'alpha':0.5}, line_kws={'color':'orange'})

# Add labels and title
plt.title("Relationship between Engine CC and Price")
plt.xlabel("Engine Size (CC)")
plt.ylabel("Price (USD)")
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
df.drop(columns=['Car_ID', 'Manufacture_Year'], inplace=True, errors='ignore')
```

```python
df_final = pd.get_dummies(df, drop_first=True)

print(df_final.corr()['Price_USD'].sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols_to_corr = ['Price_USD', 'Engine_CC', 'Mileage_km_per_l', 'Car_Age']

plt.figure(figsize=(8, 6))
sns.heatmap(df[cols_to_corr].corr(), annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Correlation Matrix of Car Features")
plt.show()
```
